```python
import numpy as np #library to handle data in vectorised form

import pandas as pd #library for data analysis
```

```python
data = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
print(len(data))
```

```python
df = data[0]
df
```

```python
df.head()
```

```python
# processing tge cells having only assigned borough
df_br = df[df.Borough != 'Not assigned']
```

```python
# combining the neighborhoods with same postalcode
df_same = df_br.groupby(['Postal Code', 'Borough'], sort = False).agg(',' .join )
```

```python
df_same.reset_index(inplace = True)
```

```python
#replacing names of neighborhood with those of borough
df_same['Neighborhood'] = np.where(df_same['Neighborhood'] == 'Not assigned', df_same['Borough'], df_same['Neighborhood'])
df_same
```

```python
df_same.shape
```

```python
geo_code = pd.read_csv("http://cocl.us/Geospatial_data") #loading csv file containing geospatial data
```

```python
geo_code.head()
```

```python
df_mr = pd.merge(df_same, geo_code)
df_mr
```

```python
df_to = df_mr[df_mr['Borough'].str.contains('Toronto', regex = False)]
df_to
```
